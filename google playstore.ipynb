```python
sns.barplot(x = pd.DataFrame(df.groupby('Airline').mean()['Price']).reset_index()['Airline'], y = pd.DataFrame(df.groupby('Airline').mean()['Price']).reset_index()['Price'])
plt.ylabel('mean_price')
plt.xlabel('airline')
plt.xticks(rotation = 90)
plt.show()
```

```python
df_Indigo = df[df['Airline']=='IndiGo']
df_Airindia = df[df['Airline']=='Air India']
df_Jet_Airways = df[df['Airline']=='Jet Airways']
df_Spice_Jet = df[df['Airline']=='SpiceJet']
df_Multiple_carriers = df[df['Airline']=='Multiple carriers']
df_GoAir = df[df['Airline']=='GoAir']
df_Vistara = df[df['Airline']=='Vistara']
df_AirAsia = df[df['Airline']=='Air Asia']
df_Vistara_Premium = df[df['Airline']=='Vistara Premium economy']
df_JetAirwaysBusisnnes = df[df['Airline']=='Jet Airways Business']
df_MultipleCarriersPremium =df[df['Airline']=='Multiple carriers Premium economy']
df_Trujet = df[df['Airline']=='Trujet']

d1 = df_Indigo.groupby('month')['Price'].mean().reset_index()
d2 = df_AirAsia.groupby('month')['Price'].mean().reset_index()
d3=df_Jet_Airways.groupby('month')['Price'].mean().reset_index()
d4=df_Spice_Jet.groupby('month')['Price'].mean().reset_index()
d5= df_Multiple_carriers.groupby('month')['Price'].mean().reset_index()
d6= df_GoAir.groupby('month')['Price'].mean().reset_index()
d7= df_Vistara.groupby('month')['Price'].mean().reset_index()
d8= df_Airindia.groupby('month')['Price'].mean().reset_index()
d9= df_Vistara_Premium.groupby('month')['Price'].mean().reset_index()
d10= df_JetAirwaysBusisnnes.groupby('month')['Price'].mean().reset_index()
d11= df_MultipleCarriersPremium.groupby('month')['Price'].mean().reset_index()
d12= df_Trujet.groupby('month')['Price'].mean().reset_index()

d1['Airline']='Indigo'
d2['Airline']='Air Asia'
d3['Airline']='Jet Airways'
d4['Airline']='SpiceJet'
d5['Airline']='Multiple carriers'
d6['Airline']='GoAir'
d7['Airline']='Vistara'
d8['Airline']='Air India'
d9['Airline']='Vistara Premium economy'
d10['Airline']='Jet Airways Business'
d11['Airline']='Multiple carriers Premium economy'
d12['Airline']='Trujet'

gt = pd.concat([d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12], axis = 0).reset_index()
sns.barplot(data = gt, x='Airline', y ='Price', hue = 'month')
plt.xticks(rotation = 90)
plt.ylabel('mean_price')
plt.show()
```
